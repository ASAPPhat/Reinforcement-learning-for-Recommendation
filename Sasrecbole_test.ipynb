```python
from recbole.quick_start import run_recbole

parameter_dict = {
    # Tắt negative sampling lúc training (chuẩn cho mô hình SASRec)
    'train_neg_sample_args': None,
    
    # Cấu hình đánh giá (Evaluation Settings)
    'eval_args': {
        'split': {'LS': 'valid_and_test'}, # Leave-one-out: train (n-2), valid (1), test (1)
        'group_by': 'user',                # Nhóm dữ liệu theo từng user
        'order': 'TO',                     # Sắp xếp theo trình tự thời gian (Temporal Order)
        'mode': 'full'                     # Bật đánh giá trên tất cả item (Full Rank)
    }
}

# Đổi tham số dataset thành 'ml-1m'
run_recbole(model='SASRec', dataset='ml-1m', config_dict=parameter_dict)
```
